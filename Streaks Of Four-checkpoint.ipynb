```python
import random

def coinToss():
    number = 10
    fliplist = ''
    
    while number > 0:
        flip = random.randint(0, 1)
        if (flip == 0):
            fliplist += '1'
        else:
            fliplist += '0'
        number -= 1 
    
    return str(fliplist)

```

```python
nope = 0
yep = 0
i = 0

while i<10000000000:
    result = coinToss()
    #print('result:' + result)

    if '1111' not in result:
        if '0000' not in result:
            nope += 1
        else:
            yep += 1
    else:
        yep += 1

    if i % 1000000 == 0:
        print('after ' + str(i) + ' trials: ' + str(yep / (nope+yep)))

    i += 1
    
```
