```python
import pandas
```

```python
# load the hospitals file
s3_hospitals = "s3://mids-w205-fall2016/hospital_compare/hospitals.csv"
df_hospitals = pandas.read_csv(s3_hospitals)

# load effective care file
s3_effective_care = "s3://mids-w205-fall2016/hospital_compare/effective_care.csv"
df_effective = pandas.read_csv(s3_effective_care)

```

```python
# check out the data
df_hospitals.head()
```

```python
# look at the state breakdown (this is an aggregation)
df_hospitals.groupby('State').agg({'Provider ID':len})
```

```python
# look at the type breakdown
df_hospitals.groupby('Hospital Type').agg({'Provider ID':len})
```

```python
# look at the emergency room breakdown
df_hospitals.groupby('Emergency Services').agg({'Provider ID':len})
```

```python
# Let's mutate the data
df_hospitals['Area Code'] = df_hospitals['Phone Number'].apply(lambda x : str(x)[0:3])
df_hospitals.head()
```

```python
# look at the effective care data
df_effective.head()
```

```python
df_effective.score2 = pandas.to_numeric(df_effective.Score, errors = 'coerce')
```

```python
# Now implement a filter
score_filter = df_effective.score2 < 100
df_effective[score_filter]
```

```python
df_effective.groupby('Provider ID').agg({'Hospital Name':len}).sort('Hospital Name', ascending=False)
```

```python
df_merge = df_hospitals.merge(df_effective, on='Provider ID')
```

```python
# Can you tell me what this does?
df_merge.groupby('Hospital Type').agg({'Provider ID':len})
```
