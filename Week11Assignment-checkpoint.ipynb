##0.1 List, Mapping, Filtering, and Reducing Comprehensions

```python

def DoesStuff(input):
    if isinstance(input, str):
        return input[::-1]
    elif isinstance(input, float) or isinstance(input, int):
        return input * 2
    else:
        return ""

fix_me = ['arsenic', 5, 'ball', 2, 47, 123, 'comprehensions', 'list', 5.5, 8343.23, 'shishkabob']

answer = [DoesStuff(stuff) for stuff in fix_me]
print("Use the function above in a list comprehension applied to the list of data")
print(answer)
```

```python
print("Do the same thing with a map.")
print(list(map(DoesStuff, fix_me)))
```

```python
# One way to filter out items < 4 using a lambda...
DoesStuffLambda = lambda x: x[::-1] if isinstance(x, str) else x*2

answer = [item for item in map(DoesStuffLambda, fix_me) if \
          ((isinstance(item, int) or isinstance(item, float)) and item >= 5) or isinstance(item, str)]

print("Do the same, but with lambdas and filter out values less than 5")
print(answer)

# Another way to do it, but using the filter function this time...
iterable = [item for item in map(DoesStuffLambda, fix_me)]
function = lambda j: j >= 5 if isinstance(j, int) or isinstance(j, float) else True

answer2 = filter(function, iterable) 

print(list(answer2))
```

## 1  Generating Random Data and Filtering That Data

```python
import numpy as np
np.random.seed(25)
ar = np.random.randn(1000)

ar = ar * 100
print("array of 1000 random numbers multiplied by 100")
print(ar)
```

```python
ar = np.int8(ar)
print("matrix of 8-bit integers")
print(ar)
```

```python
ar = np.reshape(ar, (-1, 5))
print("five columns and 200 rows")
print(ar)
```

```python
print("row wise maximums")
print(np.amax(ar, axis=1))

print("\ncolumn wise maximums")
print(np.amax(ar, axis=0))
```

```python
print("mean of the entire matrix ar")
print(np.mean(ar))
```

```python
print("number of values less than the mean")
(ar < np.mean(ar)).sum()
```

```python
print("number of values greater than mean value by row")
np.sum(ar > np.mean(ar), axis = 1)
```

```python
print("number of zeroes in above matrix")
(np.sum(ar > np.mean(ar), axis = 1) == 0).sum()
```

```python
print("sum of all values that are greater than 100 or are between 5 and -20")
ar[np.where(ar>100)].sum() + ar[np.where((ar >= -20) * (ar < 5))].sum()
```
